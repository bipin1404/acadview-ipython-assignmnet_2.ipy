```
#question 1
import numpy as np
a=np.arange(10).reshape(10,1)
a=np.random.random((10,1))
print("matrix=",a)
b=np.mean(a)
print('mean of matrix=',b)
```

```
#question 2
import numpy as np
a=np.arange(20).reshape(20,1)
a=np.random.random((20,1))
print("matrix=",a)
v=np.var(a)
s=np.std(a)
print("variance=",v)
print("standatrd deviation=",s)
```

```
#question 3
import numpy as np
a=np.random.rand(10,20)
print("matrix A=",a)
b=np.random.rand(20,25)
print("matrix B=",b)
c=np.dot(a,b)
print("matrix C=",c)
print("size of matrix c",c.size)
x=np.sum(c)
print("sum of all matrix=",x)
```
